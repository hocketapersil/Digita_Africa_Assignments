```python
import pandas as pd

# Create a DataFrame from a dictionary
data = pd.DataFrame({
    'Name': ['John', 'Alice', 'Bob', 'Emily', 'David'],
    'Age': [30, 25, 35, 28, 32],
    'City': ['New York', 'Los Angeles', 'None', 'Houston', 'San Francisco'],
    'Gender': ['M', 'F', 'M', 'F', 'M']
})

# Save DataFrame to a CSV file
data.to_csv('data.csv', index=False)

data
# Create a new DataFrame from a dictionary
new_data = pd.DataFrame({
    'Name': ['Michael', 'Sarah'],
    'Age': [41, 40],
    'City': ['Miami', 'Seattle'],
    'Gender': ['M', 'F']
})
```

## Question 7:

### How do you fill missing values in a DataFrame with a specific value?

```python
my_new_data = pd.DataFrame({'customer_name': ['Hocket', 'Pumaya', 'Ransford', 'Pascal'], 
                            'product': ['iPad Pro', 'iphone 15 pro', 'MacBook Pro', 'MacBook Air'], 
                            'Location': ['Boston', 'New York', 'Texas', 'California'],
                             'price': [10000,15000, None, 70000 ] })

my_new_data


#fill in the missing value

filled_data = my_new_data.fillna(10000)

filled_data
```

## Question 10:

### How do you perform an inner join between two DataFrames in pandas?

```python
inner_column = data.join(new_data, how='inner')
inner_column
```

## Question 11:

### How do you perform a left join between two DataFrames in pandas?

```python
left_join_column = data.join(new_data, how='inner', lsuffix='left', rsuffix='right')
left_join_column
```

## Question 12:

### How do you perform a right join between two DataFrames in pandas?

```python
right_join = data.join(new_data, how='inner', rsuffix='right', lsuffix='left')
right_join
```

## Question 13:

### How do you concatenate two DataFrames vertically in pandas?

```python
vert_concat_column = pd.concat([data, new_data], axis=0, ignore_index=True)
vert_concat_column
```

## Question 14:

### How do you concatenate two DataFrames horizontally in pandas?

```python
hor_concat_column = pd.concat([data, new_data], axis=1, ignore_index=False)
hor_concat_column
```
